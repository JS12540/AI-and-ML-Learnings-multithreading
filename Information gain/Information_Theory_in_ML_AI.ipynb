Information theory is a branch of mathematics that quantifies information

# Key Concepts in Information Theory

## Entropy (H):

Definition: Entropy is a measure of uncertainty or randomness in a probability distribution. It quantifies the average amount of information produced by a stochastic source.

Use of Information Theory in Machine Learning:

Information theory plays a significant role in machine learning in several areas:

*   Entropy as a Criterion: In decision trees, entropy is used to measure the information gain when splitting nodes. A lower entropy split (higher information gain) indicates a better split in the dataset.

*   Mutual Information for Feature Selection: Mutual information is used to evaluate the relevance of features in supervised learning. Higher mutual information between a feature and the target variable indicates higher relevance.

*   KL Divergence in Variational Inference: In variational inference (e.g., variational autoencoders), KL divergence is used to measure the difference between the approximate posterior and the true posterior, guiding the learning of probabilistic models.

Summary:
*   Entropy measures uncertainty in a distribution.
*   Mutual Information quantifies shared information between two variables.
*   KL Divergence measures how one probability distribution differs from another.

```python
import numpy as np

# Function to calculate entropy H(X)
def entropy(prob_dist):
    return -np.sum(prob_dist * np.log2(prob_dist))

# Function to calculate mutual information I(X; Y)
def mutual_information(joint_prob_dist, x_prob_dist, y_prob_dist):
    mi = 0.0
    for i in range(len(x_prob_dist)):
        for j in range(len(y_prob_dist)):
            if joint_prob_dist[i][j] > 0:  # To avoid log(0) issue
                mi += joint_prob_dist[i][j] * np.log2(joint_prob_dist[i][j] / (x_prob_dist[i] * y_prob_dist[j]))
    return mi

# Function to calculate KL Divergence D_KL(P || Q)
def kl_divergence(p_dist, q_dist):
    return np.sum(p_dist * np.log2(p_dist / q_dist))

# Example Usage:
# Entropy Example
p_x = np.array([0.5, 0.5])  # A fair coin toss (heads, tails)
print("Entropy H(X):", entropy(p_x))

# Mutual Information Example
joint_prob_xy = np.array([[0.1, 0.2], [0.2, 0.5]])  # Joint probability distribution of X and Y
p_x = np.array([0.3, 0.7])  # Marginal probability distribution of X
p_y = np.array([0.4, 0.6])  # Marginal probability distribution of Y
print("Mutual Information I(X; Y):", mutual_information(joint_prob_xy, p_x, p_y))

# KL Divergence Example
p_dist = np.array([0.4, 0.6])  # Probability distribution P
q_dist = np.array([0.5, 0.5])  # Probability distribution Q
print("KL Divergence D_KL(P || Q):", kl_divergence(p_dist, q_dist))
```
